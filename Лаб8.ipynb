```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# ÐŸÑ€Ð¸Ð¿ÑƒÑÑ‚Ð¸Ð¼Ð¾, Ñ‰Ð¾ Ñƒ Ñ‚ÐµÐ±Ðµ Ñ” Ð´Ð°Ñ‚Ð°Ñ„Ñ€ÐµÐ¹Ð¼ df Ð· Ñ‡Ð¸ÑÐ»Ð¾Ð²Ð¸Ð¼Ð¸ Ð¾Ð·Ð½Ð°ÐºÐ°Ð¼Ð¸
# Ð’Ð¸Ð±Ñ–Ñ€ ÐºÐ¾Ð»Ð¾Ð½Ð¾Ðº Ð´Ð»Ñ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¸Ð·Ð°Ñ†Ñ–Ñ— (Ð½Ð°Ð¿Ñ€Ð¸ÐºÐ»Ð°Ð´ col1, col2, ...)
x = df[['hdlngth', 'taill', 'eye', 'chest']].values
# Ð¡Ð¿Ð¸ÑÐ¾Ðº Ð´Ð»Ñ Ð·Ð±ÐµÑ€ÐµÐ¶ÐµÐ½Ð½Ñ ÑÐ¸Ð»ÑƒÐµÑ‚-ÑÐºÐ¾Ñ€Ñ–Ð²
silhouette_scores = []

# Ð”Ñ–Ð°Ð¿Ð°Ð·Ð¾Ð½ ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(x)
    score = silhouette_score(x, labels)
    silhouette_scores.append(score)

# Ð’Ñ–Ð·ÑƒÐ°Ð»Ñ–Ð·Ð°Ñ†Ñ–Ñ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ñ–Ð²
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker="o")
plt.title("ÐœÐµÑ‚Ð¾Ð´ ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð² Ð´Ð»Ñ Ð²Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ Ð¾Ð¿Ñ‚Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ k")
plt.xlabel("ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð² (k)")
plt.ylabel("Ð¡ÐµÑ€ÐµÐ´Ð½Ñ–Ð¹ ÑÐ¸Ð»ÑƒÐµÑ‚-ÑÐºÐ¾Ñ€")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

numerical_features = df.select_dtypes(include=['int64', 'float64'])
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(numerical_features)


# Ð¡Ð¿Ð¸ÑÐ¾Ðº Ð´Ð»Ñ Ð·Ð±ÐµÑ€ÐµÐ¶ÐµÐ½Ð½Ñ ÑÐ¸Ð»ÑƒÐµÑ‚-ÑÐºÐ¾Ñ€Ñ–Ð²
silhouette_scores = []

# Ð”Ð»Ñ Ð¼ÐµÑ‚Ð¾Ð´Ñƒ ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð² Ð¼Ñ–Ð½Ñ–Ð¼ÑƒÐ¼ 2 ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¸
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(x_scaled)  # fit + predict Ð¾Ð´Ð½Ð¾Ñ‡Ð°ÑÐ½Ð¾
    score = silhouette_score(x_scaled, labels)
    silhouette_scores.append(score)

# ÐŸÐ¾Ð±ÑƒÐ´Ð¾Ð²Ð° Ð³Ñ€Ð°Ñ„Ñ–ÐºÐ° ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð²
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('ÐœÐµÑ‚Ð¾Ð´ ÑÐ¸Ð»ÑƒÐµÑ‚Ñ–Ð² Ð´Ð»Ñ Ð²Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²')
plt.xlabel('ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²')
plt.ylabel('Ð¡Ð¸Ð»ÑƒÐµÑ‚Ð½Ð¸Ð¹ ÐºÐ¾ÐµÑ„Ñ–Ñ†Ñ–Ñ”Ð½Ñ‚')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# KMeans Ð´Ð»Ñ 3 ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(x_scaled)

# Ð’Ñ–Ð·ÑƒÐ°Ð»Ñ–Ð·Ð°Ñ†Ñ–Ñ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ñ–Ð²
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=x_scaled[:, 0],
    y=x_scaled[:, 1],
    hue=clusters,
    palette='viridis',
    legend='full'
)

# Ð”Ð¾Ð´Ð°Ð²Ð°Ð½Ð½Ñ Ñ†ÐµÐ½Ñ‚Ñ€Ð¾Ñ—Ð´Ñ–Ð²
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=300,
    c='red',
    label='Centroids'
)

# ÐÐ°ÑÑ‚Ñ€Ð¾ÑŽÐ²Ð°Ð½Ð½Ñ Ð³Ñ€Ð°Ñ„Ñ–ÐºÐ°
plt.title('KMeans Clustering of Features with 3 Clusters')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
